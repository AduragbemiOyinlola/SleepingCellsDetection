```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import random
from datetime import datetime, timedelta
```

```python
telco = pd.read_csv("Sleeping-Cell-KPI-Data\cell_1_KPI_Data.csv", 
                    usecols=['SDATE', 'CELL_AVAIL', 'LTE_TRAFFIC_VOL'])
telco['SDATE'] = pd.to_datetime(telco['SDATE'], format='mixed')
telco.head()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(12, 7), sharex=True)

metrics = [
    ('CELL_AVAIL', 'Average Availability (%) (15m Mean)'),
    ('LTE_TRAFFIC_VOL', 'Total LTE Traffic (15m Sum)')
]

for i, (col, title) in enumerate(metrics):
    sns.lineplot(ax=axes[i], data=telco, x='SDATE', y=col, marker='o', markersize=4)
    
    axes[i].set_title(f"{title}", fontweight='bold', fontsize=12)
    axes[i].grid(True, linestyle=':', alpha=0.7)

    y_max = telco[col].max()
    axes[i].set_ylim(0, y_max * 1.1)

    # Apply your specific Date Formatting and 2-Day Interval
    axes[i].xaxis.set_major_locator(mdates.DayLocator(interval=2)) 
    axes[i].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))

plt.tight_layout()
plt.xticks(rotation=0) 
plt.show()
```
